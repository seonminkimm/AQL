```python
start_date = '2014-05-01 00:00:00'
end_date= '2015-04-31 24:00:00'

for i in city_id:
    df_time = (globals()['city_{}_df'.format(i)]['time'] >= start_date) & (globals()['city_{}_df'.format(i)]['time'] <= end_date)
    globals()['city_{}_df'.format(i)] = globals()['city_{}_df'.format(i)][df_time]
```

```python
for i in city_id:
    globals()['city_{}_df'.format(i)]['datetime'] = globals()['city_{}_df'.format(i)]['time']
    globals()['city_{}_df'.format(i)]['datetime'] = pd.to_datetime(globals()['city_{}_df'.format(i)]['datetime'], format='%Y-%m-%d %H:%M:%S', errors='raise')
    # df = df.drop(columns = ['time'])
```

```python
for i in city_id:
    globals()['city_{}_df'.format(i)] = globals()['city_{}_df'.format(i)].set_index('datetime')
```

```python
for i in city_id:
    globals()['id_'+str(i)].sort()
    globals()['df_group_'+str(i)] = []
#     df_group = []
    for j in globals()['id_'+str(i)] :
        test = pd.DataFrame()
        test[j] = globals()['city_{}_df'.format(i)][['PM25_Concentration']][globals()['city_{}_df'.format(i)]['station_id']==j] #.
        print(i, j)
        print(test)
        globals()['df_group_'+str(i)].append(test)
```

```python
for i in city_id:
    globals()['df_all_'+str(i)] = pd.concat(globals()['df_group_'+str(i)], axis=1)
```

```python
date_range = pd.date_range('2014-05-01 00:00:00', '2015-04-30 23:00:00', freq='60T')  
for i in city_id:
    globals()['df_all_'+str(i)] = globals()['df_all_'+str(i)].reindex(index=date_range)
    globals()['df_all_'+str(i)] = globals()['df_all_'+str(i)].astype('float')
```
