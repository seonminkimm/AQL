## 이러한 센서에서 수집된 센서 데이터는 결측치가 많다.

- time-series로써 문제를 풀려면, 이러한 결측을 제거해서는 안된다.
- 결측치를 보간해야한다.
```
결측치는 일반적인 데이터에서도 어떻게 처리하는지가 정말 중요한 이슈이지만, 시계열 데이터에서는 그 중요도가 더욱 올라간다.
그 이유는 대부분의 시계열은 비정상(구간별로 다른 특성을 가지는)이기 때문에 시간에 따라 평균과 분산이 변한다.
따라서 특정 시점 값의 소실은 그 시점의 평균과 분산의 왜곡을 가져오고 결과적으로 분석 결과에 치명적인 영향을 미치게 되는 것이다.
```

- 해당 분석에서는 습도 변수에 대한 다른 변수들이 끼치는 영향을 분석하고, 관찰된 변수들로 습도를 예측하는 모델링을 함.

- 상대습도가 어떤 변수(e.g. 온도)에 따라 변화하는지 알고싶음.



> **문제**

```
▶ 공기질에 영향을 미치는 데이터들이 어떤 특성에 영향을 받는지 분석해보자.
```  

> **기대효과**

